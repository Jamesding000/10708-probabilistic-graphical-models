```python
import string
from collections import Counter
```

```python
# Parameters
vocab_file = 'data/nyt_vocab.txt'
data_file = 'data/nyt_data.txt'
new_articles = ['data/music_review_raw.txt', 'data/sport_analysis_raw.txt']
```

```python
def load_vocab(path):
    vocab = {}
    with open(path, 'r') as f:
        for i, line in enumerate(f):
            vocab[line.strip()] = i
    return vocab

def process_article(path, vocab):
    with open(path, 'r') as f:
        text = f.read().lower()
    
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = text.split()
    
    counts = Counter()
    for word in words:
        if word in vocab:
            counts[vocab[word]] += 1
    
    pairs = [f"{idx}:{cnt}" for idx, cnt in sorted(counts.items())]
    return ",".join(pairs)
```

```python
vocab = load_vocab(vocab_file)
print(f"Loaded {len(vocab)} words from vocabulary")

formatted_lines = []
for article_path in new_articles:
    print(f"Processing {article_path}...")
    formatted_lines.append(process_article(article_path, vocab))

with open(data_file, 'a') as f:
    for line in formatted_lines:
        f.write("\n" + line)

print(f"\nAppended {len(formatted_lines)} articles to {data_file}")
```
